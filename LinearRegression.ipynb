### Outlier Treatment

- Outlier present in rainfall
- Outlier present in n_hot_rooms
- Outlier present in crime_rate

```python
np.percentile(df.n_hot_rooms,[99])
uv1 = np.percentile(df.n_hot_rooms,[99])[0]
df[(df.n_hot_rooms>uv1)]
df.n_hot_rooms[(df.n_hot_rooms> 2*uv1)] = 2*uv1
```

```python
sns.jointplot(x='n_hot_rooms', y='price', data=df)
```

```python
np.percentile(df.rainfall,[99])
uv2 = np.percentile(df.rainfall,[99])[0]
df[(df.rainfall<uv2)]
df.rainfall[(df.rainfall< 3*uv2)] = 3*uv2
```

```python
sns.jointplot(x='rainfall', y='price', data=df)
```

```python
np.percentile(df.crime_rate,[99])
uv3 = np.percentile(df.crime_rate,[99])[0]
df[(df.crime_rate>uv3)]
df.crime_rate[(df.crime_rate> 3*uv3)] = 3*uv3
```

### Removing Unwanted cols

### Variable Transformation
- Object type to interger type

```python
df = pd.get_dummies(df)
df.head()
```

```python
del df['airport_NO']
del df['bus_ter_YES']
df.head()
```

# Training Model

### Test-Train Split

```python
X_multi = df.drop("price",axis=1)
y_multi = df['price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_multi,y_multi,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_multi,y_multi)
```

```python
print(model.intercept_,model.coef_)
```

```python
print(X_train.shape,X_test.shape,y_train,y_test.shape)
```

```python
y_test_model = model.predict(X_test)
y_train_model = model.predict(X_train)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train,y_train_model)
```

```python
r2_score(y_test,y_test_model)
```
